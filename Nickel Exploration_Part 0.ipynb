#Merging tables

```python
#Each BHID has interval of assay and lithology.
df1= assay.merge(lithology, how = 'inner', on = ['BHID','DEPTH_FROM','DEPTH_TO'])
df1.set_index('BHID')
```

```python
#Each BHID will have one collar and one survey.
df2= collar.merge(survey, how = 'inner', on = ['BHID','YCOLLAR','XCOLLAR','ZCOLLAR','DEPTH'])
df2.set_index('BHID')
```
